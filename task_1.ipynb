```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sales = pd.read_csv('/content/drive/MyDrive/Amazon Sale Report.csv')
sales
```

```python
#showing max columns
pd.set_option('display.max_columns', None)
sales
```

```python
#information of dataset like null values, data types, number of columns
sales.info()
```

```python
sales.columns
```

```python
sales.describe()
```

```python
#dropping duplicate values
sales.drop_duplicates(inplace= True)
sales
```

```python
#dropping columns that are of no use
sales.drop(columns=['index','Unnamed: 22'], inplace=True)
```

```python
sales
```

```python
#finding the sum of null values
sales.isnull().sum()
```

```python
# heatmap before filling for null values
sns.heatmap(sales.isnull(), cbar = False,cmap='viridis')
```

```python
#changing data type of date time
sales['Date'] = pd.to_datetime(sales['Date'], format='%m-%d-%y', errors='coerce')
```

```python
sales.info()
```

```python
#all categorical columns
a = sales.select_dtypes(include = 'object')
a
```

```python
a.isnull().sum()
```

```python
# filling null values of categorical columns by using mode
for col in ['Courier Status', 'currency', 'ship-city', 'ship-state', 'ship-country']:
    a[col].fillna(a[col].mode()[0], inplace=True)
```

```python
a.isnull().sum()
```

```python
#dropping promotion ids and fulfilled by columns becuase its consist null values more than 75%
sales.drop(columns=['promotion-ids','fulfilled-by'],inplace = True)
```

```python
sales.update(a)
sales.info()
```

```python
#filling amount column by using mean
sales['Amount'].fillna(sales['Amount'].mean(), inplace=True)
sales['Amount']
```

```python
#filling ship postal code column by using mean
sales['ship-postal-code'].fillna(sales['ship-postal-code'].mean(), inplace=True)
sales['ship-postal-code']
```

```python
sales.isnull().sum()
```

Handling outliers in numercial columns

```python
sns.boxplot(x="Amount", data=sales, palette="Set1")
plt.xlabel("Amount")
plt.show()
```

```python
Q1,Q3 = sales.Amount.quantile([0.25,0.75])
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
UL,LL
```

```python
sales.Amount = np.where(sales.Amount>UL,UL,sales.Amount)
```

```python
text_columns = ['Status', 'Fulfilment', 'Sales Channel ', 'ship-city', 'ship-state', 'ship-country', 'Category']
for col in text_columns:
    sales[col] = sales[col].str.strip().str.lower()
```
